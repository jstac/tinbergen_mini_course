## Matplotlib

* Most common plotting package in Python
* Based on MATLAB-style plotting
* Two interfaces - procedural and object-orientated

### Procedural syntax

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 200)
y = np.sin(x)

plt.plot(x, y, 'b-', linewidth=2)
plt.show()
```

### OOP syntax

```python
fig, ax = plt.subplots()
ax.plot(x, y, 'b-', linewidth=2)
plt.show()
```

More more control and more Pythonic...

Additions:

```python
fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(x, y, 'r-', linewidth=2, label='sine function', alpha=0.6)
ax.set(title='A plot')
ax.legend()
ax.grid()
plt.show()
```

Once more, with LaTex:

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 200)
y = np.sin(x)
ax.plot(x, y, 'r-', linewidth=2, label=r'$\alpha$', alpha=0.6)
ax.legend(loc='upper center')
plt.show()
```

#### Multiple plots on one axis

```python
from scipy.stats import norm
from random import uniform

fig, ax = plt.subplots()
x = np.linspace(-4, 4, 150)
for i in range(3):
    m, s = uniform(-1, 1), uniform(1, 2)
    y = norm.pdf(x, loc=m, scale=s)
    current_label = f'$\mu = {m:.2f}$'
    ax.plot(x, y, linewidth=2, alpha=0.6, label=current_label)
ax.legend()
plt.show()
```

#### Multiple axes

```python
from scipy.stats import norm
from random import uniform

num_plots = 3
colors = ['red', 'blue', 'purple']

fig, axes = plt.subplots(num_plots, 1, 
                         sharex=True, 
                         figsize=(5, 5))

x = np.linspace(-4, 4, 150)
for ax, color in zip(axes, colors):
    m, s = uniform(-1, 1), uniform(1, 2)
    y = norm.pdf(x, loc=m, scale=s)
    current_title = f'$\mu = {m:.2f}$'
    ax.plot(x, y, linewidth=2, alpha=0.6, c=color)
    ax.set(title=current_title)

fig.tight_layout()
plt.show()
```

### 3D

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm


def f(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

xgrid = np.linspace(-3, 3, 50)
ygrid = xgrid
x, y = np.meshgrid(xgrid, ygrid)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,
                y,
                f(x, y),
                rstride=2, cstride=2,
                cmap=cm.jet,
                alpha=0.7,
                linewidth=0.25)
ax.set_zlim(-0.5, 1.0)
plt.show()
```

## ipywidgets

* Package to create interactive widgets in Jupyter notebook
* Fun to use with figures
* See here for installation instructions: https://github.com/jupyter-widgets/ipywidgets#install

```python
def plot_sine(xgrid, a, b, c):
    '''
    Sine parameters
    ---------------
    a : amplitude
    b : period
    c : shift
    '''
    
    plt.plot(xgrid, a * np.sin(b * xgrid + c))
    plt.ylim(-2, 2)
    plt.show()
```

```python
xgrid = np.linspace(0, 2 * np.pi)
plot_sine(xgrid, 1, 1, 0)
```

```python
from ipywidgets import interact, interactive, fixed

interact(plot_sine, xgrid=fixed(xgrid), a=(-2, 2, 0.1), b=(-2, 2, 0.1), c=(-2, 2, 0.1))
```
